```python
import pandas as pd
from pathlib import Path
import numpy as np
import geopy
from geopy.geocoders import Nominatim

total_result = 0
counter = 0
DATA_DIR = Path('data')
```

```python
 def place_to_gps(place):
        global counter, total_result
        counter+=1
        if counter%10==True:
            print(f"{round(counter/total_result, 3)}%")
        try:
            location = geolocator.geocode(place)
            # If coordinates are found for location
            if location:
                country = location.address.rsplit(',', 1)[1]
                return pd.Series([location.latitude, location.longitude, location.address, location.altitude, location.point, country])

        except geopy.exc.GeopyError as err:
            print("geopy_exc_geopyerror")
            print(err)
            print(place)
            return pd.Series([np.nan, np.nan, np.nan, np.nan, np.nan, np.nan])
        except geopy.exc.ConfigurationError as err:
            print("geopy_exc_ConfigurationError")
            print(err)
            print(place)
            return pd.Series([np.nan, np.nan, np.nan, np.nan, np.nan, np.nan])
        except geopy.exc.GeocoderServiceError as err:
            print("geopy_exc_GeocoderServiceError")
            print(err)
            print(place)
            return pd.Series([np.nan, np.nan, np.nan, np.nan, np.nan, np.nan])
        except geopy.exc.GeocoderQueryError as err:
            print("geopy_exc_GeocoderQueryError")
            print(err)
            print(place)
            return pd.Series([np.nan, np.nan, np.nan, np.nan, np.nan, np.nan])
        except geopy.exc.GeocoderQuotaExceeded as err:
            print("geopy_exc_GeocoderQuotaExceeded")
            print(err)
            print(place)
            return pd.Series([np.nan, np.nan, np.nan, np.nan, np.nan, np.nan])
        except geopy.exc.GeocoderAuthenticationFailure as err:
            print("geopy_exc_GeocoderAuthenticationFailure")
            print(err)
            print(place)
            return pd.Series([np.nan, np.nan, np.nan, np.nan, np.nan, np.nan])
        except geopy.exc.GeocoderInsufficientPrivileges as err:
            print("geopy_exc_GeocoderInsufficientPrivileges")
            print(err)
            print(place)
            return pd.Series([np.nan, np.nan, np.nan, np.nan, np.nan, np.nan])
        except geopy.exc.GeocoderTimedOut as err:
            print("geopy_exc_GeocoderTimedOut")
            print(err)
            print(place)
            return pd.Series([np.nan, np.nan, np.nan, np.nan, np.nan, np.nan])
        except geopy.exc.GeocoderUnavailable as err:
            print("geopy_exc_GeocoderUnavailable")
            print(err)
            print(place)
            return pd.Series([np.nan, np.nan, np.nan, np.nan, np.nan, np.nan])
        except geopy.exc.GeocoderParseError as err:
            print("geopy_exc_GeocoderParseError")
            print(err)
            print(place)
            return pd.Series([np.nan, np.nan, np.nan, np.nan, np.nan, np.nan])
        except geopy.exc.GeocoderNotFound as err:
            print("geopy_exc_GeocoderNotFound")
            print(err)
            print(place)
            return pd.Series([np.nan, np.nan, np.nan, np.nan, np.nan, np.nan])
        except:
            print("unknown error")
            return pd.Series([np.nan, np.nan, np.nan, np.nan, np.nan, np.nan])

        return pd.Series([np.nan, np.nan, np.nan, np.nan, np.nan, np.nan])
```

```python
dfs = {}
for fname in DATA_DIR.glob('*.pkl'):
    print(f'Reading data from {fname}')
    dfs[fname.name.split('_')[0]] = pd.read_pickle(fname)

    
    df = pd.concat(dfs.values(), axis=0)
    result = df['user.location'].mask(df['user.geo_enabled'], other=np.nan)
    geolocator = Nominatim(user_agent="tapp")
    
    counter = 0
    total_result = len(result)
    df[['geo_latitude', 'geo_longitude', 'geo_address', 'geo_altitude', 'geo_point', 'geo_country']] = result.apply(place_to_gps)

    print(df)

    df.to_pickle(f'geo_{fname}.pkl')
```
